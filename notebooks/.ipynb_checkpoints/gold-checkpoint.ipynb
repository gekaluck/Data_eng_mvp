```python
df_snaphots = spark.table("silver.crypto.price_snapshots").where("snapshot_date >= DATE '2026-03-19' AND snapshot_date < DATE '2026-03-21'").orderBy("coin_id", "snapshot_date").drop("change_percent_24hr")
df_snaphots.toPandas()
# df_snaphots_w_coins_info = df_snapshots.merge(coins, df_snaphots.coin_id == coins.id, how="left)
```

```python
from pyspark.sql.window import Window
import pyspark.sql.functions as F

windowSpec = Window.partitionBy("coin_id").orderBy("snapshot_date")


df_with_diff = df_snaphots.withColumn(
    "prev_price", F.lag("price_usd", 1).over(windowSpec)
).withColumn(
    "price_change_pct", (F.col("price_usd") - F.col("prev_price")) / F.col("prev_price") * 100
)

df_with_diff = df_with_diff.where("prev_price is not null")
df_with_diff.toPandas()
```
